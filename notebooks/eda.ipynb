```python
# ðŸ“Š Exploratory Data Analysis (EDA)
# Este notebook descarga y analiza los datos de COVID-19 
# para Argentina, Chile, Bolivia, Paraguay, Brasil y Uruguay.

# Importar librerÃ­as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Definir la URL del dataset
URL = "https://covid.ourworldindata.org/data/owid-covid-data.csv"

# Descargar los datos
df = pd.read_csv(URL)

# Filtrar por paÃ­ses de interÃ©s
countries = ["ARG", "CHL", "BOL", "PRY", "BRA", "URY"]
df = df[df["iso_code"].isin(countries)]

# Guardar los datos localmente en una carpeta oculta
df.to_csv("../.data/covid_data.csv", index=False)

print("Datos descargados y guardados correctamente.")

```

```python
# Convertir la columna 'date' a formato datetime
df["date"] = pd.to_datetime(df["date"])

# Agrupar por mes y paÃ­s, sumando los casos totales diarios
df_monthly = df.groupby([df["date"].dt.to_period("M"), "iso_code"])["new_cases"].sum().reset_index()
df_monthly["date"] = df_monthly["date"].dt.to_timestamp()  # Convertir periodos a timestamps

# Definir estilos de grÃ¡fico
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Graficar la evoluciÃ³n de casos mensuales por paÃ­s
sns.lineplot(data=df_monthly, x="date", y="new_cases", hue="iso_code", marker="o", linewidth=2)

# Formato del eje X para mostrar menos etiquetas
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Solo mostrar 10 fechas en el eje X

# TÃ­tulos y etiquetas
plt.title("COVID-19: Nuevos Casos Mensuales por PaÃ­s", fontsize=14)
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Nuevos Casos", fontsize=12)
plt.legend(title="PaÃ­s")

# Mostrar grÃ¡fico
plt.show()
```

```python
# Convertir la columna 'date' a formato datetime
df["date"] = pd.to_datetime(df["date"])

# Ordenar datos
df = df.sort_values(["iso_code", "date"])

# Calcular un promedio mÃ³vil de 7 dÃ­as para suavizar las variaciones diarias
df["new_cases_smooth"] = df.groupby("iso_code")["new_cases"].transform(lambda x: x.rolling(7, min_periods=1).mean())

# Filtrar los datos para evitar valores extremos (opc)
df = df[df["new_cases_smooth"] > 0]

# Definir estilo del grÃ¡fico
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Graficar con el promedio mÃ³vil en lugar de los valores brutos
sns.lineplot(data=df, x="date", y="new_cases_smooth", hue="iso_code", linewidth=2)

# Espaciado del eje X para mejor lectura
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Solo mostrar 10 fechas en el eje X

# TÃ­tulos y etiquetas
plt.title("COVID-19: Nuevos Casos Suavizados por PaÃ­s (Promedio MÃ³vil 7 dÃ­as)", fontsize=14)
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Nuevos Casos Promedio", fontsize=12)
plt.legend(title="PaÃ­s")

# Mostrar grÃ¡fico
plt.show()
```
